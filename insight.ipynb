```python
import re

def clean_column_name(col):
    col = str(col).strip().lower()
    col = col.replace(" ", "_")
    col = re.sub(r"[^a-z0-9_]", "", col)
    col = re.sub(r"_+", "_", col)
    return col

df.columns = [clean_column_name(col) for col in df.columns]

print(df.columns.tolist())
```

```python
rename_map = {
    "source_platform": "platform",
    "current_price_idr": "price",
    "original_price_idr": "original_price",
    "seller_name": "seller",
    "seller_rating": "seller_rating"
}

df = df.rename(columns=rename_map)

print(df.columns.tolist())
```

```python
print("Shape sebelum filter harga:", df.shape)

if "price" in df.columns:
    df = df[df["price"] >= 10_000_000].copy()

print("Shape sesudah filter harga:", df.shape)
display(df.head())
```

```python
import pandas as pd
import numpy as np
import re

# =====================================
# GABUNGIN KOLOM SOLD TANPA bfill WARNING
# =====================================
candidate_cols = [col for col in df.columns if "sold_count" in col.lower()]
print("Kolom kandidat sold:", candidate_cols)

if len(candidate_cols) > 0:
    temp = df[candidate_cols[0]].copy()
    for col in candidate_cols[1:]:
        temp = temp.combine_first(df[col])
    df["sold_count_clean"] = temp
else:
    df["sold_count_clean"] = np.nan


# =====================================
# FUNGSI CLEAN SOLD YANG BENAR
# =====================================
def clean_sold(val):
    if pd.isna(val):
        return 0

    val = str(val).strip().lower()

    if val in ["", "nan", "none", "null", "-", "--"]:
        return 0

    # hapus kata dulu
    val = val.replace("terjual", "")
    val = val.replace("sold", "")
    val = val.replace("+", "")
    val = val.strip()

    # kasus 1: format rb / k, misal 1.2rb atau 1,2rb
    match_rb = re.search(r"(\d+[.,]?\d*)\s*(rb|k)", val)
    if match_rb:
        num = match_rb.group(1).replace(",", ".")
        return int(float(num) * 1000)

    # kasus 2: format ribuan biasa, misal 2,397 atau 1.234
    if re.fullmatch(r"[\d.,]+", val):
        # kalau ada koma dan titik sekaligus, ambil angka saja
        # kalau cuma koma / titik sebagai pemisah ribuan, juga ambil angka saja
        digits_only = re.sub(r"[^\d]", "", val)
        return int(digits_only) if digits_only != "" else 0

    # kasus 3: campuran teks + angka biasa
    match_num = re.search(r"\d+", val)
    if match_num:
        return int(match_num.group())

    return 0


# =====================================
# APPLY
# =====================================
df["sold_count"] = df["sold_count_clean"].apply(clean_sold)

print(df[["sold_count_clean", "sold_count"]].head(15))
print("\nNaN sold_count:", df["sold_count"].isna().sum())
print("Jumlah 0 terjual:", (df["sold_count"] == 0).sum())
print("\nContoh nilai sold terbesar:")
print(df["sold_count"].sort_values(ascending=False).head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if "brand" not in df.columns:
    raise ValueError("Kolom 'brand' tidak ditemukan di dataframe.")

df["brand"] = df["brand"].astype(str).str.strip()
df["brand"] = df["brand"].replace(["nan", "None", "none", "", "-", "--"], "Unknown")
df["brand"] = df["brand"].str.title()

print(df["brand"].value_counts().head(20))
```

```python
brand_counts = df["brand"].value_counts().head(15)

plt.figure(figsize=(12, 6))
plt.bar(brand_counts.index, brand_counts.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 15 Brand dengan Jumlah Produk Terbanyak")
plt.xlabel("Brand")
plt.ylabel("Jumlah Produk")
plt.tight_layout()
plt.show()
```

```python
def classify_brand(brand):
    brand = str(brand).lower()

    if brand in ["apple"]:
        return "Apple"
    elif brand in ["asus", "msi", "alienware", "razer"]:
        return "Gaming/Premium"
    elif brand in ["lenovo", "hp", "dell", "acer"]:
        return "Mainstream"
    else:
        return "Other"

df["brand_category"] = df["brand"].apply(classify_brand)

print(df["brand_category"].value_counts())
```

```python
cat_counts = df["brand_category"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(cat_counts.index, cat_counts.values)
plt.title("Distribusi Kategori Brand")
plt.xlabel("Kategori Brand")
plt.ylabel("Jumlah Produk")
plt.tight_layout()
plt.show()
```

```python
brand_sold_summary = (
    df.groupby("brand", dropna=False)
      .agg(
          jumlah_produk=("brand", "count"),
          total_terjual=("sold_count", "sum"),
          rata_rata_terjual=("sold_count", "mean"),
          rata_rata_harga=("price", "mean")
      )
      .sort_values(by="total_terjual", ascending=False)
)

display(brand_sold_summary.head(15))
```

```python
import matplotlib.pyplot as plt

top_brand_sold = brand_sold_summary.head(15)

plt.figure(figsize=(12,6))
plt.bar(top_brand_sold.index, top_brand_sold["total_terjual"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 15 Brand Berdasarkan Total Produk Terjual")
plt.xlabel("Brand")
plt.ylabel("Total Terjual")
plt.tight_layout()
plt.show()
```

```python
top_avg_sold = (
    df.groupby("brand", dropna=False)["sold_count"]
      .mean()
      .sort_values(ascending=False)
      .head(15)
)

plt.figure(figsize=(12,6))
plt.bar(top_avg_sold.index, top_avg_sold.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 15 Brand Berdasarkan Rata-rata Produk Terjual")
plt.xlabel("Brand")
plt.ylabel("Rata-rata Terjual per Listing")
plt.tight_layout()
plt.show()
```

```python
import re
import numpy as np

def classify_laptop_category(title):
    if pd.isna(title):
        return "Unknown"

    title = str(title).lower()

    # gaming
    gaming_keywords = [
        "gaming", "rog", "tuf", "legion", "predator", "omen", "victus",
        "nitro", "katana", "cyborg", "raider", "strix", "zephyrus",
        "loq", "aorus", "helios", "g series", "fx506", "fx507"
    ]

    # business / office
    business_keywords = [
        "thinkpad", "latitude", "elitebook", "probook", "travelmate",
        "expertbook", "vostro", "dynabook"
    ]

    # creator / workstation
    creator_keywords = [
        "zenbook pro", "vivobook pro", "creator", "workstation", "zbook",
        "precision", "xps", "macbook pro", "studio"
    ]

    # casual / daily
    casual_keywords = [
        "ideapad", "aspire", "vivobook", "inspiron", "pavilion",
        "swift", "matebook", "gram", "envy", "macbook air"
    ]

    for kw in gaming_keywords:
        if kw in title:
            return "Gaming"

    for kw in business_keywords:
        if kw in title:
            return "Business"

    for kw in creator_keywords:
        if kw in title:
            return "Creator/Workstation"

    for kw in casual_keywords:
        if kw in title:
            return "Casual/Daily"

    return "Other"

df["laptop_category"] = df["title"].apply(classify_laptop_category)

print(df["laptop_category"].value_counts())
```

```python
for cat in df["laptop_category"].unique():
    print(f"\n=== {cat} ===")
    display(df[df["laptop_category"] == cat][["title", "brand", "sold_count", "price"]].head(5))
```

```python
category_sales_summary = (
    df.groupby("laptop_category")
      .agg(
          jumlah_produk=("laptop_category", "count"),
          total_terjual=("sold_count", "sum"),
          rata_rata_terjual=("sold_count", "mean"),
          rata_rata_harga=("price", "mean")
      )
      .sort_values(by="total_terjual", ascending=False)
      .reset_index()
)

display(category_sales_summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(category_sales_summary["laptop_category"], category_sales_summary["total_terjual"])
plt.title("Kategori Laptop Terlaris Berdasarkan Total Penjualan")
plt.xlabel("Kategori Laptop")
plt.ylabel("Total Terjual")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
